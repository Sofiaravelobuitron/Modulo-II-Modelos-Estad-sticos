# **Variables aleatorias.**

## Introducción
En probabilidad y estadística, el concepto de **variable aleatoria** resulta fundamental para poder modelar fenómenos inciertos.  
Una variable aleatoria puede entenderse como una función que asigna un número real a cada resultado posible de un experimento aleatorio.  
De esta manera, es posible traducir sucesos abstractos en valores numéricos que permiten el análisis matemático.

Existen dos tipos de variables aleatorias:

- **Discretas**: aquellas que toman un número finito o contable de valores (por ejemplo, el número de siniestros reportados en un año, el lanzamiento de un dado o el número de clientes que llegan a una oficina en un día).  
- **Continuas**: aquellas que pueden tomar infinitos valores dentro de un intervalo real (por ejemplo, el tiempo de espera en un hospital, la estatura de una persona o el monto de un siniestro médico).

Para cada variable aleatoria se definen funciones asociadas que describen su comportamiento probabilístico:  
- La **función de probabilidad (pmf)** en el caso discreto.  
- La **función de densidad (fdp)** en el caso continuo.  
- La **función de distribución acumulada (cdf)**, común a ambos tipos.  

Estas funciones permiten calcular probabilidades de eventos, obtener medidas de tendencia central como la **esperanza**, medir la dispersión mediante la **varianza** y derivar otras propiedades útiles.  
Además, existen funciones generadoras de momentos (FGM) y de probabilidades (FGP), que son herramientas analíticas para derivar propiedades adicionales de una distribución.

El estudio de las variables aleatorias constituye la base para comprender la **estadística inferencial**, ya que nos permite modelar la incertidumbre y sentar las bases para la estimación de parámetros, construcción de intervalos de confianza y pruebas de hipótesis que se abordan en los siguientes temas.

## Variable aleatoria
Consideremos que tenemos un experimento aleatorio cualquiera, junto con un espacio de probabilidad asociado $(\Omega,\mathcal{F},\mathbb{P})$.\
Una **variable aleatoria** es una tranformación $X$ del espacio de resultados $\Omega$ al conjunto de número reales, esto es,
$$X:\Omega\rightarrow\mathbb{R},$$
tal que para cualquier número real x
$$\{\omega \in \Omega :X(\omega)\leq x\} \in \mathcal{F}$$

Considerando el conjunto de valores que una variable aleatoria puede tomar,
vamos a clasificar a las variables aleatorias en dos tipos: discretas o continuas.


### Variable Aleatoria Discreta.
Decimos que una variable aleatoria es discreta cuando el conjunto de valores que ésta toma es un conjunto discreto, es decir, un conjunto finito o numerable.\
Por ejemplo, el conjunto $\{0,1,2,\cdots,n)\}$ es un conjunto discreto porque es finito, lo mismo con $\mathbb{N}$, aunque es infinito es numerable y en consecuencia discreto. 

### Variable Aleatoria Continua
Una variable aleatoria es continua cuando toma todos los valores dentro de un intervalo $(a,b) \subseteq \mathbb{R}$

## Función de probabilidad
Sea $X$ una variable aleatoria discreta con valores $x_0,x_1,\cdots$. La función de probabilidad de $X$, denotada por $f(x) : \mathbb{R}\rightarrow\mathbb{R}$, se define como sigue
$$ 
 f(x)=
 \left \{
 \begin{aligned}
 &\mathbb{P}(X=x) &\quad\text{si } x=x_0,x_1\cdots \\
 &0  &\quad\text{en otro caso } 
 \end{aligned}
 \right.$$
 La función de probabilidad es simplemente la función que indica la probabilidad en los distintos valores que toma la variable aleatoria.

## Función de densidad
Sea $X$ una variable aleatoria continua. Decimos que la función integrable y no negativa $f(x):\mathbb{R}\rightarrow \mathbb{R}$ es la funciín de densidad de $X$ si para cualquier intervalo $[a,b]$ de $\mathbb{R}$ se cumple la igualdad 
$$\mathbb{P}(X \in [a,b])=\int_a^b f(x)dx$$

## Función de distribución.
Sea $X$ una variable aleatoria cualquiera. La funciín de distribución de $X$, denotada por $F(x):\mathbb{R}\rightarrow\mathbb{R}$, se define como la probabilidad 
$$F(x)=\mathbb{P}(X\leq x)$$
La función de distribución ebañuada en un número x cualquiera es la pprobabilidad de que la variable aleatoria tome un valor menor o igual a $x$, en otras palabras, tome un valor en el intervalo $(-\infty,x]$

Una variable aleatoria $X$ es continua si su función de distribución $F(x)$ es una función continua.\
Toda función de ditribució $F(x)$ satisface las siguientes propiedades:\
a) $\lim\limits_{x\to\infty}F(x)=1$\
b) $\lim\limits_{x\to\infty}F(x)=0$\
c) Si $x_1\leq x_2$, entonces $F(x_1)\leq F(x_2)$\
d) $F(x)=F(x+)$\
Toda función $F(x):\mathbb{R}\rightarrow\mathbb{P}$ que cumpla las $4$ propiedades se le llama función de distribución, sin tener necesariamente una variable aleatoria que la defina.

## Independencia de variables aleatorias.
Se dice que las variables aleatorias $X$ y $Y$ son independientes si los eventos $(X\leq x)$ y $(Y\leq y)$ son independientes para cualesquiera valores reales $x$ y $y$, es decir, si se cumple la igualdad 
$$\mathbb{P}[(X\leq x)\cap(Y\leq y)]=\mathbb{P}(X\leq x)\mathbb{P}(Y\leq y)$$
- Caso discreto
$$\mathbb{P}(X=x,Y=y)=\mathbb{P}(X=x)\mathbb{P}(Y=y), \quad -\infty<x,y<\infty$$
- Caso continuo
$$f_{X,Y}(x,y)=f_X(x)f_Y(y),  \quad -\infty<x,y<\infty$$

## Esperanza.
Sea $X$ una variable aleatoria discreta con función de probabilidad $f(x)$, La esperanza de $X$ se define como el número 
$$\mathbb{E}(X)=\sum_{x}xf(x)$$
suponiendo que esta suma es absolutamente convergente, es decir, cuando la suma de los valores absolutos es convergente.\
Si $X$ es continua con función de densidad $f(x)$, entonces la esperanza es 
$$\mathbb{E}(X)=\int_{-\infty}^{\infty}{xf(x)dx},$$
suponiendo que la integral es absolutamente convergente, es decir, cuando la integral de los valores absolutos es convergente.

Sea $X$ una variable aleatoria discreta y sea $\phi$ una función tal que $\phi(X)$ es una variable aleatoria con esperanza finita.\
Entonces
$$\mathbb{E}[\varphi(X)]=\sum_{x}\varphi(x)\mathbb{P}(X=x)$$.

Sea $X$ una variable aleatoria discreta y sea $\phi$ una función tal que $\phi(X)$ es una variable aleatoria con esperanza finita.\
Entonces
$$\mathbb{E}[\varphi(X)]=\int_{-\infty}^{\infty}\varphi(x)f(x)dx$$.

## Varianza
Sea $X$ una variable aleatoria discreta con función de probabilidad $f(x)$. La varianza de $X$ se define con el número 
$$Var(X)=\sum_{x}(x-\mu)^2f(x)$$
cuando esta suma es convergente y en donde $\mu$ es la esperanza de $X$.\
Para una variable aleatoria continua $X$ con función de densidad $f(x)$ se define
$$Var(X)=\int_{-\infty}^{\infty}(x-\mu)^2f(x)dx,$$
cuando esta integral es convergente.

## Momentos
Los momentos de una variable aleatoria $X$ o de una distribución son la colección de números:
$$\mathbb{E}(X),\mathbb{E}(X^2),\mathbb{E}(X^3),\cdots$$
correspondientes al primer momentos, segundo momentos, etc., cuando tales cantidades existan.
Para las variables aleatorias discretas, el $n$-ésimo momento se calcula mediante la siguiente fórmula
$$\mathbb{E}(X^n)=\sum_{n}f(x)$$
Para variables aleatorias continuas la fórmula es 
$$\mathbb{E}(X^n)=\int_{-\infty}^{\infty}x^nf(x)dx$$

## Cuantiles 
Sea $p \in (0,1]$, El cuantil $p$ de una variable aleatoria o de su función de distribución $F(x)$ es el número $c_p$ más pequeño, cuando existe, tal que
$$F(c_p\geq) p$$

## Moda
La moda de una variable aleatoria $X$ o de su distribución es aquel número $x*$, cuando existe, en donde la función de densidad o de probabilidad $f(x)$ toma su vañor máximo.


## Función generadora de probabilidad
Sea $X$ una variable aleatoria discreta con posibles valores del conjunto $\{0,1,2,\cdots\}$. A la función $G(t)$, definida como aparece abajo, se le llama la función generadora de probabilidad de $X$,
$$ G(t)=\mathbb{E}(t^X)=\sum_{x=0}^{\infty}t^x\mathbb{P}(X=x)$$


Sea $X$ una variable aleatoria discreta con valores en el conjunto $\{0,1,\cdots\}$ y con función generadora de probabilidad $G(t)$.\
Para $x=0,1,\cdots$
$$ \mathbb{P}(X=x)=\frac{1}{x!}G^{(x)}(0)$$

Sea $X$ una variable aleatoria discreta con función generadora de probabilidad $G(t)$, Si el $k$-ésimo momento de $X$ existen entonces 
$$G^{(x)}(1-)\mathbb{E}(X(X-1)\cdots(X-k+1))$$

Sean $X$ y $Y$ dos variables aleatorias discretas e independientes con funciones generadoras de probabilidad $G_X(t)$ y $G_Y(t)$.\
Entonces
$$G_{X+Y}(t)=G_X(t)G_Y(t)$$.

Sean $X$ y $Y$ dos variables aleatorias discretas con el mismo conjunto de valores $\{0,1,2,\cdots\}$ y con funciones generadoras de probabilidad $G_X(t)$ y $G_Y(t)$, tales que $G_X(t)=G_Y(t)$ para $t\in(-s,s)$ con $s>0$, Entonces $X$ y $Y$ tienen la misma distribución de probabilidad.

## Función generadora de momentos.
La función generadora de momentos de una variable aleatoria discreta o continua $X$ es la función $M(t)$ definida como sigue
$$ M(t)=\mathbb{E}(e^{tX})$$
para los valores reales de $t$ en donde esta esperanza existe.

Para el caso discreto esta función se calcula como 
$$M(t)=\sum_{x}e^{tx}\mathbb{P}(X=x)$$
para el caso continuo
$$M(t)=\int_{-\infty}^{\infty}e^{tx}f(x)dx$$

## Ley de los grandes números
Sea $X_1,X_2,\cdots$ una suceción infinita de variables aleatorias independientes e indénticamente distribuidas con medida finita $\mu$. entonces, cuando $n\rightarrow \infty$,
$$\frac{1}{n}\sum_{i=1}^{n}X_i\rightarrow \mu$$
en donde la cinvergencia se verifica en el sentido casi seguro (ley fuerte) y también en probabilida (ley débil).

## Teorema central del límite
Sea $X_1,X_2,\cdots$ una sucesión infinita de variables aleatorias independientes e idénticamente distribuidas, con media $\mu$ y varianza finita $\sigma^2$. entonces la función de distribución de la variable aleatoria 
$$Z_n=\frac{(X_1+\cdots+X_n)-n\mu}{\sqrt{n\sigma^2}}$$
tiende a la función de distribución normal estándar cuando $n$ tiende a infinito.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

plt.style.use("seaborn-v0_8")
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
axes = axes.flatten()

# Discretas

# Bernoulli
p = 0.3
x = [0,1]
axes[0].bar(x, st.bernoulli.pmf(x, p), color="c")
axes[0].set_title("Bernoulli(p=0.3)")

# Binomial
n, p = 10, 0.5
x = np.arange(0, n+1)
axes[1].stem(x, st.binom.pmf(x, n, p))
axes[1].set_title("Binomial(n=10, p=0.5)")

# Geométrica
p = 0.4
x = np.arange(1, 15)
axes[2].stem(x, st.geom.pmf(x, p))
axes[2].set_title("Geométrica(p=0.4)")

# Poisson
lam = 3
x = np.arange(0, 15)
axes[3].stem(x, st.poisson.pmf(x, lam))
axes[3].set_title("Poisson(λ=3)")

# Ejes
for ax in axes:
    ax.set_xlabel("x")
    ax.set_ylabel("Probabilidad / Densidad")

plt.tight_layout()
plt.show()
```

### Caso practico.

Una fábrica produce focos y cada foco tiene una probabilidad de $0.1$ de salir defectuoso. Se selecciona un lote de 20 focos. Sea $X$ el número de focos defectuosos en el lote.
- Calcular $\mathbb{P}(X\leq 2)$, es decir haya máximo 2 defectuosos.
- $\mathbb{E}[X]=np=20(0.1)=2$
- $Var[X]=np(1-p)=20(0.1)(0.9)=1.8$

```python
import scipy.stats as st

# Parámetros
n, p = 20, 0.1
X = st.binom(n, p)

# Probabilidad
print("P(X ≤ 2) =", round( X.cdf(2),6))

# Esperanza y varianza
print("E[X] =", X.mean())
print("Var[X] =", X.var())
```

### Caso practico
La estatura de los estudiantes de una universidad se modela con una distribución normal de media 170 cm y desviación estándar 10 cm. Sea $Y$ la altura de un estudiante elegido al azar.
- Calcular $\mathbb{P}(160\leq Y\leq 180)$
- $\mathbb{E}[Y]$
- $Var[Y]$

```python


# Parámetros
mu, sigma = 170, 10
Y = st.norm(mu, sigma)

# Probabilidad
p = Y.cdf(180) - Y.cdf(160)
print("P(160 ≤ Y ≤ 180) =", round( p,6))

# Esperanza y varianza
print("E[Y] =", Y.mean())
print("Var[Y] =", Y.var())
```
