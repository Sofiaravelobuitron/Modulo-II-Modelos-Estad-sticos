- **Enfoque geométrico**: útil cuando los eventos se representan en intervalos continuos o figuras geométricas.

## Experimento aleatorio
**Experimento Aleatorio**: es un experimento $\epsilon$ que reproducido en condiciones idénticas puede conducir a varios resultados, en el cual no se puede predecir el resultado.

## Espacio muestral
El espacio de todos los posibles resultados de un experimento aleatorios es llamdo **espacio muestral**, y es denotado por $\Omega$. Un posible resultado del experimento aleatorio se denota por $\omega$, donde $\omega \in \Omega$.
- El lanzamiento de una moneda: $\Omega = ${águila, sol}.
- El lanzamiento de un dado : $\Omega ={1,2,3,4,5,6}$
- El lanzamiento de un dardo en un circulo de 30cm de diámetro: $\Omega ={(x,y):\sqrt{x^2 +y^2} \leq 15}$
- Romedo espera a Julieta, el cual va desde 1 minuto a 1 hora : $\Omega =[0,1]$.
- Cuando estudiamos el tiempo de vida de una bacteria : $\Omega =[0, \infty)$

```python
import itertools

# Ejemplo: Lanzamiento de dos monedas
monedas = ["Águila", "Sol"]
espacio_muestral = list(itertools.product(monedas, repeat=2))
print("Espacio muestral (Ω):", espacio_muestral)
print("Número de resultados posibles:", len(espacio_muestral))
```

**Evento Aleatorio**: Es un subconjutno de $\Omega$, donde se puede decir si el experimento se realiza o no. Un evento aleatorio $A$ es un sucbconjunto de $\Omega$.
- A: Se obtine sol al lanzar una moneda.
- A: Se obtiene un 3 al lanzar un dado.
- Julieta se retrasa menos de un cuarto de hora : $ A = [0,\frac{1}{4}]$

Así, un evento aleatorio esta representado por un conjunto de resultados ára los cuales este se realiza.

- **Potencia**: conjunto de todos los subconjuntos de $\Omega$.
- **Evento cierto** : $\emptyset$
- Si A un evento, entonces $ A^c$ (no sucede A).
- $A \cap B$: sucede A y B.
- $A \cup B$: sucede A o B.

**Espacio de Probabilidad**: es una terna $(\Omega,\mathcal{F}, \mathbb{B})$, donde $\Omega$ es un conjunto no vacío, $\mathcal{F}$ es la familia de subconjuntod de $\Omega$ y $\mathbb{P}$ es una medida de probabilidad, la cual le asigna un número no negativo a un evento, está ultima es una función $\mathbb{P}:\mathcal{F} \rightarrow [0,1]$ que verifica los siguientes axiomas:
- Si $A\in \mathcal{F}$, entonces $0 \leq \mathbb{P}(A)\leq 1$
- La probabilidad del espacio de estados $\Omega$ es igual a 1, es decir, $\mathbb{P}=1$
- Si $\left(A_n\right)_{n\geq}$ es una sucesión finita o infinita numerable de elementos de $\mathcal{F}$, que son mutuamente excluyentes dos a dos, es decir, $A_i\cap A_j = \emptyset$ para $i \neq j$, entonces \
$\begin{equation}  \mathbb{P} \left(\bigcup_{n=1}^{\infty} A_n\right)= \sum_{n=1}^{\infty}\mathbb{P}(A_n)
\end{equation}$


Propiedades:
- $\mathbb{P}(\emptyset)=0$
- $\mathbb{P}(A^c)=1-\mathbb{P}(A)$
- $A \subset B \Rightarrow \mathbb{P}(A)\leq\mathbb{P}(B)$
- $A\subset B \Rightarrow \mathbb{P}(B \setminus A)=\mathbb{P}(B)-\mathbb{P}(A)$
- $\mathbb{P}(A\cup B)=\mathbb{P}(A)+\mathbb{P}(B)-\mathbb{P}(A\cap B)$

## Probabilidad Axiomática
La fundamentacion teórica de la probabilidad se dio mediante una construcción axiomática, la cual se debe a Andrei Kolmogorov y fué publicada en el año de 1933. Esta relaciona la teoría de la probabilidad con la teoría de la medida, la cual era reciente y fue trabajada por Emile Borel y Henri Lebesgue.


En la definición axiomática de la probabildad no se define la manera explícita de calcular probabilidades, sino únicamente se proponen las reglas que el cálculo de probabilidades debe satisfacer.
Recordemos que un axioma es una proposición que se acepta como válida y sobre la cual se funda una teoría.
**Axiomas de probabilidad**
1. $\mathbb{P}(A)\neq 0$.
2. $\mathbb{P}(\Omega)=1$
3. $\mathbb{P}(\bigcup_{k=1}^\infty A_k)=\sum_{k=1}^\infty\mathbb{P}(A_k)\quad \text{ cuando } A_1,A_2,\cdots \text{ son ajenos dos a dos.}$

A cualquier función $\mathbb{P}$ definida sobre una colección de eventos satisfaga los tres axiomas de Kolmogorov, se le llama medida de probabilidad o simplemente probabilidad.

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Definimos dos eventos disjuntos A y B
venn2(subsets=(1,1,0), set_labels=('A', 'B'))
plt.title("Eventos disjuntos: P(A ∪ B) = P(A) + P(B)")
plt.show()
```

**Ejemplo**
Supongamos que lanzamos un dado:  

- Sea $(A = \{2,4\})$, la probabilidad es $(\mathbb{P}(A) = \frac{2}{6} =\frac{1}{3})$.  
- Sea $(B = \{6\})$, la probabilidad es $(\mathbb{P}(B) = \frac{1}{6})$.  

Como $A$ y $B$ son disjuntos, se cumple:  


$$P(A \cup B) = P(A) + P(B) = \frac{1}{3} + \frac{1}{6} = \frac{1}{2}$$



## Probabilidad Clásica (Laplaciano)
La definición clásica de probabilidad fue introducida en el año 1812por Pierre-Simon Laplace, se basó en el concepto de equprobabilidad como propiedad objetiva de los eventos asociados a un experimentos aleatorio.
Supongamos que $\Omega = \{\omega_1,\omega_2,...,\omega_n\}$ es un subconjunto finito no vacío y sea $\mathcal{F}=\mathbb{P}(\Omega)$, los elemtos $\{\omega_i\}$ tienen la misma probabilidad $\frac{1}{n}$ y se tiene que la probabilidad de un evento A es:\
$$\begin{equation}
\mathbb{P}(A)=\frac{\text{casos favorables}}{\text{casos totales}}=\frac{|A|}{|\Omega|}, \text{para toda} A\subset \Omega
\end{equation} $$

Consideremos el experimento de lanzar un dado de seis caras, El espacio muestral esta formado por los números de 1 al 6. \
Sea el evento $A=$"obtener un número par". \
Como todos los resultados son equiprobables, la probabilidad se calcula con el cociente:
$$\mathbb{P}(A)=\frac{|A|}{|\Omega|}=\frac{3}{6}=0.5$$

```python
# Probabilidad clásica: dado
omega = {1,2,3,4,5,6}
A = {2,4,6}  # evento: número par

P_A = len(A)/len(omega)
print("P(A) =", P_A)
```

## Probabilidad Geométrica
En este modelo se busca encontrar probabilidades de la siguiente manera:
$$\mathbb{P}(A)=\frac{\text{área deseada}}{\text{área total}}$$
Ejemplos:
- Problema del encuentro (Romeo y Julieta)
- Aguja de Buffon.

Supongamos que elegimos un número real al azar en el intervalo $[0,1]$. Sea el evento $A=$"el número elegido es menor que 0.3". \
En este caso, la probabilidad se determina como la proporción de longuitudes:
$$ \mathbb{P}(A)=\frac{\text{longuitud de intervalo }A}{\text{longuitud del intervalo total}}=\frac{0.3}{1}=0.3$$
Lo verificaremos mediante una simulación.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación: número aleatorio uniforme en [0,1]
n = 100000
muestra = np.random.rand(n)

# Evento A: número < 0.3
A = muestra[muestra < 0.3]

P_simulada = len(A)/n
P_teorica = 0.3

print("Probabilidad simulada:", P_simulada)
print("Probabilidad teórica:", P_teorica)

# Visualización
plt.hist(muestra, bins=50, color="skyblue", edgecolor="black")
plt.axvspan(0, 0.3, color="red", alpha=0.3, label="Evento A: x<0.3")
plt.legend()
plt.title("Modelo Geométrico: Uniforme en [0,1]")
plt.show()
```

**Problema del encuentro**
Dos personas acuerdan encontrarse en un lugar entre las 8:00 y las 9:00 de la mañana. Cada una llega a una hora elegida al azar y de manera independiente en ese intervalo. La primera que llega espera 15 minutos como máximo; si en ese lapso no llega la otra, se va.
¿Cuál es la probabilidad de que se encuentren?\
representamos los tiempos de llegada como variables aleatorias uniformes.
El espacio muestral es el cuadrado $[0,60]\times [0,60]$, de áres total $60^2=3,600$\
Se encuentran si $|X-Y|\leq 15$, es decir la diferencia entre sus horas d ellegada es menor o igual a $15$ minutos.\
Gráficamente, corresponde a la franja alrededor de la diagonal $X=Y$, limitada por las rectas $Y=X+15$ y $Y=X-15$.\
El área de la franja es: 
$$ A=3600-2\cdot \frac{(60-15)^2}{2}=3600-2\cdot 2025/2=3600-2025=1575$$
La probabilidad de encuentro es:
$$\mathbb{P}=\frac{1575}{3600}=\frac{7}{16}\approx 0.4375$$



```python
import matplotlib.pyplot as plt
import numpy as np
# Rango del tiempo [0,60]
x = np.linspace(0, 60, 500)
y1 = x + 15
y2 = x - 15

plt.figure(figsize=(6,6))
plt.fill_between(x, y1, y2, where=(y1<=60) & (y2>=0), color="skyblue", alpha=0.5, label="Encuentro")
plt.plot(x, x, 'k--', label="Y=X")
plt.plot(x, y1, 'c', label="Y = X+15")
plt.plot(x, y2, 'm', label="Y = X-15")

plt.xlim(0,60)
plt.ylim(0,60)
plt.xlabel("Llegada Persona 1 (minutos)")
plt.ylabel("Llegada Persona 2 (minutos)")
plt.title("Problema del Encuentro")
plt.legend()
plt.grid(True)
plt.show()
```

## Probabilidad Frecuentista

Sea $n_A$ el número de ocurrencias de un evento $A$ en $n$ realizaciones de un experimento aleatorio. La probabilidad frecuentista del evento $A$ se define como el límite 

$$P(A)=\lim\limits_{n \to \infty}\frac{n_A}{n}$$

Consideramos el experimento de lanzar una moneda:
- 1 lanzamiento: 1 cara. $\leftarrow$ No podemos decir nada.
- 3 lanzamientos: 2 caras y 1 sol. $\leftarrow$ No podemos decir nada.
- 300 lanzamientos: 200 caras y 100 soles.$\leftarrow$ Empezamos a hacernos preguntas.
- 3000 lanzamientos :2000 caras y 1000 soles.$\leftarrow$ es casi seguro que la moneda está cargada.

La repetición del experimento es la que nos da la infomación, y tenemos que la frecuencia empírica del evento cara es $\frac{2000}{3000}=\frac{2}{3}$.\
Esta frecuencia nos da una idea de la posibilidad de obtener cara en el lazamiento de la moneda porque el número de preticiones es grande.

Lanzamos una moneda equilibrada un número grande de veces y registramos la frecuencia relativa de que aparezca “cara”. Según el modelo frecuentista y la ley de los grandes números, conforme aumenta el número de lanzamientos, la frecuencia relativa debería aproximarse a la probabilidad teórica $0.5$.\
En la siguiente simulación mostramos cómo evoluciona la frecuencia relativa de “cara” conforme incrementamos el número de lanzamientos.

```python
import random
import matplotlib.pyplot as plt

n = 10000
resultados = [random.choice(["Cara","Cruz"]) for _ in range(n)]

# Frecuencia relativa de cara
freq_cara = [sum(1 for r in resultados[:k] if r=="Cara")/k for k in range(1,n+1)]

print("Frecuencia relativa final de cara:", freq_cara[-1])

# Gráfico de convergencia
plt.plot(range(1,n+1), freq_cara, color="c", label="Frecuencia relativa")
plt.axhline(0.5, color="m", linestyle="--", label="Probabilidad teórica")
plt.xlabel("Número de lanzamientos")
plt.ylabel("Frecuencia relativa")
plt.title("Convergencia de la probabilidad")
plt.legend()
plt.show()
```

## Probabilidad Condicional
Sean $A$ y $B$ dos eventos supongamos que $B$ tiene probabilidad estrictamente positiva. La probabilidad condicional del evento $A$, dado el evento $B$, se denota por el símbolo $\mathbb{P}(A|B)$ y se define como el conciente \
$$ \mathbb{P}(A|B)=\frac{\mathbb{P}(A\cap B)}{\mathbb{P}(B)}
$$

Consideremos el experimento de lanzar dos dados. Definimos:
- Evento $A$: "la suma de los dos dados es par".
- Evento $B$: "al menos uno de los dados muestra un 6"
La probabilidad condicional $\mathbb{P}(A|B)$ se calcula con la fórmula:
$$\mathbb{P}(A|B)=\frac{\mathbb(A\cap B)}{\mathbb{P}(B)}$$
Usamos una simulacióm para aproximar este valor.

```python
import random

n = 100000
dados = [ (random.randint(1,6), random.randint(1,6)) for _ in range(n) ]

# Evento A: suma par
A = [d for d in dados if (sum(d)%2==0)]

# Evento B: al menos un 6
B = [d for d in dados if (d[0]==6 or d[1]==6)]

P_A = len(A)/n
P_B = len(B)/n
P_AyB = len([d for d in dados if (sum(d)%2==0 and (d[0]==6 or d[1]==6))])/n
P_A_dado_B = P_AyB / P_B

print("P(A) =", P_A)
print("P(B) =", P_B)
print("P(A∩B) =", P_AyB)
print("P(A|B) ≈", round(P_A_dado_B,6))
```

## Teorema de Bayes
Sea $B_1,\cdots,B_n$ una partición de $\Omega$ tal que $\mathbb{P}\neq 0, i =1,\cdots,n.$Sea $A$ un evento tal que $\mathbb{P}(A)\neq 0$. Entonces para cada $j=1,2,\cdots,n$, \
$$\mathbb{P}(B_j|A)=\frac{\mathbb{P}(A|B_j)\mathbb{P}(B_j)}{\sum_{i=1}^n \mathbb{P}(A|B_i)\mathbb{P}(B_i)}$$

Un test médico tiene la siguientes carcterísticas:
- El $1%$ de la población padece la enfermedad.
- El test da positivo el $95 \%$ de los enfermos (sensibilidad).
- El test da positivo el $5\%$ de los sanos (falsos positivos).
Si una persona recibe un resultado positivo, ¿cuál es la probabilidad de que realmente esté enferma?

```python
# Ejemplo Bayes: prueba médica
P_enfermo = 0.01
P_positivo_dado_enfermo = 0.95
P_positivo_dado_sano = 0.05

# Probabilidad total de positivo
P_positivo = (P_enfermo * P_positivo_dado_enfermo) + ((1-P_enfermo) * P_positivo_dado_sano)

# Bayes
P_enfermo_dado_positivo = (P_enfermo * P_positivo_dado_enfermo) / P_positivo
print("P(enfermo | positivo) =", round(P_enfermo_dado_positivo,6))
```

## Independia de Eventos
Se dice que los eventos $A$ y $B$ son independientes si se cumple la igualdad 
$$ \mathbb{P}(A\cap B) = \mathbb{P}(A)\mathbb{P}(B)$$

Decimos que $n$ eventos $A_1,\cdots,A_n$ son (mutuamente) independientes si se satisfacen todas y cada una de las condiciones siguientes:
$$
\begin{align}
\mathbb{P}(A_i \cap A_j)&= \mathbb{P}(A_i)\mathbb{P}(A_j),\quad i,j \text{ distintos}\\
\mathbb{P}(A_i \cap A_j\cap A_k)&= \mathbb{P}(A_i)\mathbb{P}(A_j)\mathbb{P}(A_k), \quad i,j,k\text{ distintos}\\
&\vdots\\
\mathbb{P}(A_1 \cdots\cap A_n)&= \mathbb{P}(A_1)\cdots\mathbb{P}(A_n)
\end{align}
$$

Se dice que un colección infinita de eventos es independiente si cualquier subcolección finita de ella lo es.

Supongamos el experimento de lanzar un dado. Definimos:
- Evento $A$: "obtner un número par"
- Evento $B$: "obtener un número mayor que 3".
Para verificar si son independientes, comparamos:
$$\mathbb{P}(A\cap B) \text{ vs } \mathbb {P}(A)\cdot \mathbb {P}(B)$$
Si ambos valores coinciden, los eventos son independientes, de lo contrario, no lo son.

```python
# Ejemplo independencia
omega = {1,2,3,4,5,6}

A = {2,4,6}  # número par
B = {4,5,6}  # número mayor que 3

P_A = len(A)/len(omega)
P_B = len(B)/len(omega)
P_AyB = len(A & B)/len(omega)

print("P(A) =", P_A, "P(B) =", P_B, "P(A∩B) =", P_AyB)
print("P(A)*P(B) =", P_A*P_B)

if abs(P_A*P_B - P_AyB) < 1e-6:
    print(" A y B son independientes")
else:
    print(" A y B NO son independientes")
```
